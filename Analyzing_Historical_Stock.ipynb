```python
# Question 1 - Extracting Tesla Stock Data Using yfinance

import yfinance as yf
```

```python
# Create a Ticker object for Tesla (TSLA)
tesla_ticker = yf.Ticker("TSLA")
```

```python
# Extract Tesla stock data
tesla_stock_data = tesla_ticker.history(period="max")
```

```python
# Display the first five rows of Tesla stock data
print("Tesla Stock Data:")
print(tesla_stock_data.head())
```

```python
# Question 2 - Extracting Tesla Revenue Data Using Webscraping

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# Define the URL for Tesla revenue data
tesla_revenue_url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla-inc/revenue"
```

```python
# Send an HTTP request to the webpage
tesla_revenue_html = requests.get(tesla_revenue_url).text
```

```python
# Parse the HTML content
soup_tesla_revenue = BeautifulSoup(tesla_revenue_html, 'html.parser')
```

```python
# Define headers to mimic a user agent
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Send an HTTP request to the webpage with headers
tesla_revenue_html = requests.get(tesla_revenue_url, headers=headers).text

# Parse the HTML content
soup_tesla_revenue = BeautifulSoup(tesla_revenue_html, 'html.parser')

# Find the table with the specified class name
tesla_revenue_table = soup_tesla_revenue.find("table", class_="historical_data_table")

# Extract Tesla revenue data and store it in a DataFrame
tesla_revenue_data = pd.read_html(str(tesla_revenue_table))[0]

# Display the first five rows of Tesla revenue data
print("\nTesla Revenue Data:")
print(tesla_revenue_data.head())
```

```python
# Display the last five rows of Tesla revenue data
print("Tesla Revenue Data:")
print(tesla_revenue_data.tail())
```

```python
# Rename the columns
tesla_revenue_data.columns = ['Date', 'Revenue']

# Display the first five rows of Tesla revenue data with updated column names
print("\nTesla Revenue Data:")
print(tesla_revenue_data.head())
```
